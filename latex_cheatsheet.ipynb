# LaTeX in Jupyter — Cheat Sheet

Jupyter renders LaTeX math inside **Markdown cells** using MathJax.  
Use `$...$` for **inline** math and `$$...$$` for **display** (centered) math.

---
## 1. Inline vs Display Mode

**Inline** — math flows with text:  
The eigenvalue equation is $A\vec{v} = \lambda\vec{v}$, where $\lambda$ is a scalar.

**Display** — math is centered on its own line:

$$A\vec{v} = \lambda\vec{v}$$

---
## 2. Vectors

Arrow notation: $\vec{v}$, $\vec{u}$, $\vec{w}$

Bold notation: $\mathbf{v}$, $\mathbf{u}$, $\mathbf{w}$

Hat (unit vector): $\hat{i}$, $\hat{j}$, $\hat{k}$

Column vector:

$$\vec{v} = \begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix} = \begin{bmatrix} 3 \\ -1 \\ 7 \end{bmatrix}$$

Row vector:

$$\vec{v}^T = \begin{bmatrix} 3 & -1 & 7 \end{bmatrix}$$

---
## 3. Matrices

A 2×2 matrix:

$$A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$$

A 3×3 matrix:

$$B = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}$$

Matrix with parentheses instead of brackets:

$$C = \begin{pmatrix} a & b \\ c & d \end{pmatrix}$$

Augmented matrix (with a vertical bar):

$$\left[\begin{array}{cc|c} 1 & 2 & 5 \\ 3 & 4 & 6 \end{array}\right]$$

## 8. Multi-line Equations (aligned)

Use `\begin{aligned}` to align equations at the `&` symbol:

$$
\begin{aligned}
2x + 3y &= 7 \\
4x - y  &= 1
\end{aligned}
$$

More complex alignment:

$$
\begin{aligned}
\det(A) &= ad - bc \\
        &= (1)(4) - (2)(3) \\
        &= 4 - 6 \\
        &= -2
\end{aligned}
$$

## 9. Putting It All Together — Example

### Solving $A\vec{x} = \vec{b}$

Given:

$$
A = \begin{bmatrix} 2 & 1 \\ 5 & 3 \end{bmatrix}, \quad
\vec{b} = \begin{bmatrix} 4 \\ 7 \end{bmatrix}
$$

We find $\vec{x} = A^{-1}\vec{b}$. First compute the inverse:

$$
A^{-1} = \frac{1}{\det(A)} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
= \frac{1}{6-5} \begin{bmatrix} 3 & -1 \\ -5 & 2 \end{bmatrix}
= \begin{bmatrix} 3 & -1 \\ -5 & 2 \end{bmatrix}
$$

Then:

$$
\vec{x} = \begin{bmatrix} 3 & -1 \\ -5 & 2 \end{bmatrix} \begin{bmatrix} 4 \\ 7 \end{bmatrix}
= \begin{bmatrix} 5 \\ -6 \end{bmatrix}
$$

```python
# Verify the example above with NumPy
import numpy as np

A = np.array([[2, 1],
              [5, 3]])

b = np.array([4, 7])

x = np.linalg.solve(A, b)
print(f"x = {x}")  # Should be [5, -6]
```

## 💡 Tips

1. **Double-click** a rendered Markdown cell to see/edit the raw LaTeX source.
2. **Shift + Enter** to render the cell and move to the next.
3. Use `\quad` or `\qquad` to add horizontal spacing: $a \quad b \qquad c$
4. Use `\text{...}` for regular text inside math: $\text{if } x > 0$
5. Use `\color{red}{...}` to color parts: $\color{red}{\vec{v}}$ + $\color{blue}{\vec{w}}$
