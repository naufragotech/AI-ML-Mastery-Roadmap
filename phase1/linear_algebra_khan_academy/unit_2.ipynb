## Unit 2: Matrix transformations

### Lesson 1: Functions and linear transformations

Some definitions:

- Function $\rightarrow$ A mapping from a domain to a codomain

- Domain $\rightarrow$ Set which the functions can be applied to

- Codomain $\rightarrow$ Set which the function maps to

- Range $\rightarrow$ Subset of the codomain that the function actually maps to

- Transformation $\rightarrow$ Function operating on vectors

#### Linear Transformation

A Transformation $T: \reals^n \rightarrow \reals^m$ is a Linear Transformation **if and only if**:

- $T(\vec{a} + \vec{b}) = T(\vec{a}) + T(\vec{b}) $

- $T(\lambda \vec{a}) = \lambda T(\vec{a})$

A linear transformation can be represented as a matrix-vector multiplication:

$$ T: \reals^n \rightarrow \reals^m $$

$$ T(\vec{x}) = \mathbf{A} \vec{x}

#### Identity Matrix

Given a matrix in $\reals^{n \times n}$, it is called identity if all diagonal elements are 1, and all the others are 0. Such as:

$$ I_n = \begin{bmatrix} 1 & 0 & 0 & \dots & 0 \\ 0 & 1 & 0 & \dots & 0 \\ 0 & 0 & 1 & \dots & 0 \\ \vdots & \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & 0 & \dots & 1 \end{bmatrix} $$

Main property of the Identity Matrix:

$$ I_n \cdot \vec{x} = \vec{x} $$

#### Standard Basis

The column vectors of the identity matrix are called the standard basis for $\reals^n$

$$ I_n = \begin{bmatrix} \vec{e_1} & \vec{e_2} & \dots & \vec{e_n} \end{bmatrix} $$

$$ \left\{ \vec{e_1}, \vec{e_2}, \dots, \vec{e_n}  \right\}\text{ standard basis for } \reals^n $$

When applying a linear transformation, we can also represent it like this:

$$ T(\vec{x}) = \begin{bmatrix} T(\vec{e_1}) & T(\vec{e_2}) & \dots & T(\vec{e_n}) \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} $$

#### Image of a Transformation

$V$ is a subsapce in $\reals^n$, then:

$$ T(V) \text{: image of }V \text{ under }T $$

It's also called the Image of the Transformation:

$$ T(V) = im(T) $$

And as any transformation can be represented as a matrix - vector multiplication, the column space of the matrix is equivalent to the image:

$$ C(A) = span(\vec{a_1}, \vec{a_2}, \dots, \vec{a_n}) = im(T) $$

#### Preimage of a Transformation

Given supaces $A$ and $S$, and spaces $X$ and $Y$, such as:

$$ T: X \rightarrow Y $$

$$ A \subset X \space \mid \space S \subset Y $$

$$ T\begin{pmatrix}A\end{pmatrix} \rightarrow S $$

Then, the preimage of $S$ under $T$ is noted as:

$$ T^{-1}\begin{pmatrix}S\end{pmatrix} \text{: preimage of } S \text{ under } T $$

$$ T^{-1}\begin{pmatrix}S\end{pmatrix} = A = \left\{ \vec{x} \in X \space \mid \space T\begin{pmatrix}\vec{x}\end{pmatrix} \in S \right\} $$

$$ T\begin{pmatrix}T^{-1}\begin{pmatrix}S\end{pmatrix}\end{pmatrix} = S $$

#### Kernel of a Transformation

Is the subset of vectors that when applied the transformation, give the zero vector $\vec{0}$

$$ Ker\begin{pmatrix}T\end{pmatrix} \text{: Kernel of }T = \left\{ \vec{x} \in \reals^n \mid T\begin{pmatrix}\vec{x}\end{pmatrix} = \vec{0} \right\} $$

Is the same concept as the null space of the matrix associated with the transformation

#### Operations with Transformations

Given:

$$ S: \reals^n \rightarrow \reals^m \quad \mid \quad T: \reals^n \rightarrow \reals^m $$

$$ \begin{pmatrix}S + T\end{pmatrix}\begin{pmatrix}\vec{x}\end{pmatrix} = S\begin{pmatrix}\vec{x}\end{pmatrix} + T\begin{pmatrix}\vec{x}\end{pmatrix} $$

$$ \begin{pmatrix} \lambda S \end{pmatrix} \begin{pmatrix} \vec{x} \end{pmatrix} = \lambda \begin{pmatrix} S \begin{pmatrix} \vec{x} \end{pmatrix} \end{pmatrix} $$

Also, representing the transformation as matrix:

$$ S \begin{pmatrix} \vec{x} \end{pmatrix} = \mathbf{A} \vec{x} \quad \mid \quad T \begin{pmatrix} \vec{x} \end{pmatrix} = \mathbf{B} \vec{x} $$

$$ \mathbf{A} \in \reals^{m \times n} \quad \mid \quad \mathbf{B} \in \reals^{m \times n} $$

$$ \mathbf{A} = \begin{bmatrix} \vec{a_1} & \vec{a_2} & \dots & \vec{a_n} \end{bmatrix} \quad \mid \quad \mathbf{B} = \begin{bmatrix} \vec{b_1} & \vec{b_2} & \dots & \vec{b_n} \end{bmatrix} $$

Then, the sum of the transformations can be represented as:

$$ \begin{pmatrix} S + T \end{pmatrix} \begin{pmatrix} \vec{x} \end{pmatrix} = S \begin{pmatrix} \vec{x} \end{pmatrix} + T \begin{pmatrix} \vec{x} \end{pmatrix} $$

$$ \begin{pmatrix} S + T \end{pmatrix} \begin{pmatrix} \vec{x} \end{pmatrix} = \mathbf{A} \vec{x} + \mathbf{B} \vec{x} $$

$$ \begin{pmatrix} S + T \end{pmatrix} \begin{pmatrix} \vec{x} \end{pmatrix} = \begin{pmatrix} \mathbf{A} + \mathbf{B} \end{pmatrix} \vec{x} $$

$$ \mathbf{A} + \mathbf{B} = \begin{bmatrix} \vec{a_1} + \vec{b_1} & \vec{a_2} + \vec{b_2} & \dots & \vec{a_n} + \vec{b_n} \end{bmatrix} $$

Finally, the scalar multiplication of the transformations can be represented as:

$$ \begin{pmatrix} \lambda S \end{pmatrix} \begin{pmatrix} \vec{x} \end{pmatrix} = \lambda \begin{pmatrix} S \begin{pmatrix} \vec{x} \end{pmatrix} \end{pmatrix} $$

$$ \begin{pmatrix} \lambda S \end{pmatrix} \begin{pmatrix} \vec{x} \end{pmatrix} = \lambda \mathbf{A} \vec{x} $$

$$ \lambda A = \begin{bmatrix} \lambda \vec{a_1} & \lambda \vec{a_2} & \dots & \lambda \vec{a_n} \end{bmatrix} $$
