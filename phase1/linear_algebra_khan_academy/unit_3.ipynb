## Unit 3: Alternate coordinate systems (bases)

### Lesson 1: Orthogonal complements

#### Orthogonal complement

Given a subspace $V \subseteq \reals^n$, then its orthogonal complement $V^\perp$ is:

$$ V^\perp = \left\{ \vec{x} \in \reals^n \, \mid \, \vec{x} \cdot \vec{v} = 0 \, \mid \, \forall \, \vec{v} \in V \right\} $$

Additionally, $V^\perp$ is also a subspace of $\reals^n$, so $\forall \, \vec{a}, \vec{b} \in V^\perp$:

- $\vec{a} + \vec{b} \in V^\perp $

- $\lambda \vec{a} \in V^\perp $

If the set $\left\{ \vec{v_1}, \vec{v_2}, \dots, \vec{v_k} \right\}$ is a basis for the subspace $V$, then the following apply:

- $V = \text{span}(\vec{v_1}, \vec{v_2}, \dots, \vec{v_k}) $

- $\dim(V) = k$

We define a matrix $A = \begin{bmatrix} \vec{v_1} & \vec{v_2} & \dots & \vec{v_n} \end{bmatrix}$, which column space is the subspace $V$, therefore:

- $\text{rank}(A) = \text{rank}(A^T) = k$

- $N(A^T) = C(A)^\perp = V^\perp$

- $dim(V^\perp) = dim(N(A^T)) = \text{nullity of } A^T$

- $\text{rank}(A) + \text{nullity}(A) = \text{rank}(A^T) + \text{nullity}(A^T) = n$

- $\dim(V) + \dim(V^\perp) = n$

#### Intersection between a subspace and its orthogonal complement

Given a subspace $V \subseteq \reals^n$ and its orthogonal complement $V^\perp \subseteq \reals^n$:

$$ V \cap V^\perp = \left\{ \vec{0} \right\}

#### Basis of a subspace and its orthogonal complement

Given:

$$ V \subseteq \reals^n \quad \mid \quad \dim(V) = k \quad \mid \quad V = span(\vec{v_1}, \vec{v_2}, \dots, \vec{v_k}) $$

$$ V^\perp \subseteq \reals^n \quad \mid \quad \dim(V^\perp) = n -k \quad \mid \quad V^\perp = span(\vec{w_1}, \vec{w_2}, \dots, \vec{w_{n - k}}) $$

$$ \therefore \text{The set } \left\{ \vec{v_1}, \vec{v_2}, \dots, \vec{v_k}, \vec{w_1}, \vec{w_2}, \dots, \vec{w_{n - k}} \right\} \text{ is a basis for } \reals^n $$

Finally:

$$ \forall \, \vec{x} \in \reals^n \quad \mid \quad \exists! \, \vec{v} \in V \quad \text{and} \quad \exists! \, \vec{w} \in V^\perp \quad \mid \quad \vec{x} = \vec{v} + \vec{w} $$

#### Orthogonal complement of the orthogonal complement

Given a subspace $V \subseteq \reals^n$ and its orthogonal complement $V^\perp \subseteq \reals^n$, then:

$$ (V^\perp)^\perp = V $$

#### Orthogonal complements of the null space, left null space, column space, and row space

Given a matrix $A \in \reals^{m \times n}$, the following is true:

- The orthogonal complement of the row space is the null space: $C(A^T)^\perp = N(A)$

- The orthogonal complement of the null space is the row space: $N(A)^\perp = C(A^T)$

- The orthogonal complement of the column space is the left null space: $C(A)^\perp = N(A^T)$

- The orthogonal complement of the left null space is the column space: $N(A^T)^\perp = C(A)$

#### Vectors in the row space as solutions to inhomogeneus equations

For a matrix $A \in \reals^{m \times n}$, we have:

$$ A\vec{x} = \vec{0} \quad \text{where} \quad \vec{x} \in N(A) $$

$$ A\vec{x} = \vec{b} \quad \text{where} \quad \vec{x} \in \reals^n $$

$$N(A) \subseteq \reals^n \quad \rightarrow \quad N(A)^\perp \subseteq \reals^n \quad \rightarrow \quad C(A^T) = N(A)^\perp $$

$$ \Rightarrow \exists! \, \vec{n_0} \in N(A) \quad \text{and} \quad \exists! \, \vec{r_0} \in C(A^T) \quad \mid \quad \vec{x} = \vec{r_0} + \vec{n_0} $$

$$ \therefore \vec{r_0} \text{ is a solution of } A \vec{x} = \vec{b} $$

In other words, there exists a unique vector in the row space that is a solution for each vector in the column space and have the smallest length:

$$ \forall \space \vec{b} \in C(A) \space \rightarrow \space \exists! \space \vec{r_0} \in C(A^T) \space \mid \space A \vec{r_0} = \vec{b} \quad \text{and} \quad \|\vec{x}\| \geq \|\vec{r_0}\| $$

### Lesson 2: Orthogonal projections

#### Projections

Given a line $L = \left\{ \lambda \vec{v} \space \mid \space \lambda \in \reals \right\}$, the projection of the vector $\vec{x}$ is:

$$ Proj_L(\vec{x}) = \frac{\vec{x} \cdot \vec{v}}{\vec{v} \cdot \vec{v}} \vec{v} $$

Given a subspace $V \subseteq \reals^n$ and its orthogonal complement $V^\perp \subseteq \reals^n$, then:

$$ \vec{x} \in \reals^n \quad \rightarrow \quad \vec{x} = \vec{v} + \vec{w} \quad \mid \quad \vec{x} \in V \quad , \quad \vec{w} \in V^\perp $$

$$ Proj_V\vec{x} = \vec{v} \quad \mid \quad Proj_{V^\perp}\vec{x} = \vec{w} $$

Given a subspace $V \subseteq \reals^n$ associated with the matrix $A$ such that $A \vec{x} = \vec{b} \space , \space \vec{b} \in V$ where all column vectors of $A$ form the basis for $V$.The projection can be computed as:

$$ Proj_V\vec{x} = A (A^T A)^{-1} A^T \vec{x} $$



#### Example: Projection onto a Subspace

**Problem:**
Find the projection of the vector $\vec{x}$ onto the subspace $V$. The basis for the subspace $V$ is formed by $\vec{v_1}$ and $\vec{v_2}$, and the vector $\vec{x}$ is given as:

$$
\vec{v_1} = \begin{bmatrix} 1 \\ 0 \\ 0 \\ 1 \end{bmatrix}, \quad
\vec{v_2} = \begin{bmatrix} 0 \\ 1 \\ 0 \\ 1 \end{bmatrix}, \quad
\vec{x} = \begin{bmatrix} 1 \\ 2 \\ 3 \\ 4 \end{bmatrix}
$$

**Theory Recall:**
To find the projection of a vector $\vec{x}$ onto a subspace $V$, we construct a matrix $A$ where columns are the basis vectors of $V$: $A = \begin{bmatrix} \vec{v_1} & \vec{v_2} \end{bmatrix}$. The projection is given by:

$$
Proj_V \vec{x} = A(A^T A)^{-1} A^T \vec{x}
$$

**Step-by-Step Solution:**

First, construct the matrix $A$ and compute $A^T A$:

$$
A = \begin{bmatrix} 1 & 0 \\ 0 & 1 \\ 0 & 0 \\ 1 & 1 \end{bmatrix} \quad \rightarrow \quad
A^T = \begin{bmatrix} 1 & 0 & 0 & 1 \\ 0 & 1 & 0 & 1 \end{bmatrix}
$$

$$
A^T A = \begin{bmatrix} 1 & 0 & 0 & 1 \\ 0 & 1 & 0 & 1 \end{bmatrix} \begin{bmatrix} 1 & 0 \\ 0 & 1 \\ 0 & 0 \\ 1 & 1 \end{bmatrix} = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
$$

Next, find the inverse of $A^T A$:

$$
\det(A^T A) = (2)(2) - (1)(1) = 3
$$

$$
(A^T A)^{-1} = \frac{1}{3} \begin{bmatrix} 2 & -1 \\ -1 & 2 \end{bmatrix}
$$

Now, compute $A^T \vec{x}$:

$$
A^T \vec{x} = \begin{bmatrix} 1 & 0 & 0 & 1 \\ 0 & 1 & 0 & 1 \end{bmatrix} \begin{bmatrix} 1 \\ 2 \\ 3 \\ 4 \end{bmatrix} = \begin{bmatrix} 5 \\ 6 \end{bmatrix}
$$

Then, compute $(A^T A)^{-1} A^T \vec{x}$:

$$
(A^T A)^{-1} A^T \vec{x} = \frac{1}{3} \begin{bmatrix} 2 & -1 \\ -1 & 2 \end{bmatrix} \begin{bmatrix} 5 \\ 6 \end{bmatrix} = \frac{1}{3} \begin{bmatrix} (2)(5) + (-1)(6) \\ (-1)(5) + (2)(6) \end{bmatrix} = \frac{1}{3} \begin{bmatrix} 4 \\ 7 \end{bmatrix} = \begin{bmatrix} 4/3 \\ 7/3 \end{bmatrix}
$$

Finally, compute the projection $Proj_V \vec{x} = A(A^T A)^{-1} A^T \vec{x}$:

$$
Proj_V \vec{x} = \begin{bmatrix} 1 & 0 \\ 0 & 1 \\ 0 & 0 \\ 1 & 1 \end{bmatrix} \begin{bmatrix} 4/3 \\ 7/3 \end{bmatrix} = \begin{bmatrix} 4/3 \\ 7/3 \\ 0 \\ 11/3 \end{bmatrix}
$$

$$
\boxed{Proj_V \vec{x} = \begin{bmatrix} 4/3 \\ 7/3 \\ 0 \\ 11/3 \end{bmatrix}}
$$

```python
import numpy as np

# Define vectors
v1 = np.array([1, 0, 0, 1])
v2 = np.array([0, 1, 0, 1])
x = np.array([1, 2, 3, 4])

# Define matrix A
A = np.column_stack((v1, v2))

# Compute elements
A_T = A.T
ATA = A_T @ A
ATA_inv = np.linalg.inv(ATA)
ATx = A_T @ x

# Compute projection
proj_x = A @ (ATA_inv @ ATx)

# Print intermediate steps and final projection
print(f"A^T A =\n{ATA}")
print(f"\n(A^T A)^-1 =\n{ATA_inv}")
print(f"\nA^T x = {ATx}")
print(f"\nProj_V x = {proj_x}")

# Verification
expected_proj = np.array([4/3, 7/3, 0, 11/3])
print(f"\n✅ Verified: {np.allclose(proj_x, expected_proj)}")
```

**Interpretation:**
The projected vector $Proj_V \vec{x}$ represents the closest point in the subspace $V$ to the original vector $\vec{x}$. Geometrically, the error vector $\vec{e} = \vec{x} - Proj_V \vec{x}$ corresponds to the orthogonal component of $\vec{x}$ with respect to $V$, proving it lies in the left null space $N(A^T)$.

#### Closest distance to a subspace

The closes distance from a vector to a subspace is always the projection:

$$ \|\vec{x} - \text{Proj}_V\vec{x}\| \leq \| \vec{x} - \vec{b} \| \quad \text{where} \quad \vec{x} \in \reals^n \space , \space V \subseteq \reals^n \space , \space \vec{b} \in V $$

#### Least Square approximation

The least square approximation of a vector $\vec{b}$ in a subspace $V$ which has no solution to $A\vec{x} = \vec{b}$, is noted as $\vec{x}^*$. To find it we can do the following:

$$ A\vec{x}^* = \text{Proj}_V\vec{b} \quad \rightarrow \quad \vec{x}^* = (A^T A)^{-1} A^T \vec{b} $$

### Lesson 3: Change of basis

#### Basis of a subspace

Given the basis $B$ of the subspace $V$, any vector $\vec{a} \in V$ can be represented as a linear combination of the vectors of the basis $B$. This is:

$$ B = \left\{ \vec{v_1}, \vec{v_2}, \dots, \vec{v_k} \right\} \quad \rightarrow \quad \begin{bmatrix} \vec{a} \end{bmatrix}_B = \begin{bmatrix} \lambda_1 \\ \lambda_2 \\ \vdots \\ \lambda_k \end{bmatrix} \quad \rightarrow \quad \vec{a} = \lambda_1 \vec{v_1} + \lambda_2 \vec{v_2} + \dots + \lambda_k c_k $$

#### Change of basis matrix

$C$ is called a "change of basis matrix" if:

$$ C \in \reals^{n \times k} \quad \mid \quad C = \begin{bmatrix} \vec{v_1} & \vec{v_2} & \dots & \vec{v_k} \end{bmatrix} \quad \mid \quad C \begin{bmatrix} \vec{a} \end{bmatrix}_B = \vec{a} $$

If $C$ is invertible, or $span(B) = \reals^n$, then:

$$ \begin{bmatrix} \vec{a} \end{bmatrix}_B = C^{-1} \vec{a} $$

#### Transformation matrix with respect to a basis

Give a transformation $T$, and a basis $B$, then:

$$ T: \reals^n \rightarrow \reals^n \quad \mid \quad T(\vec{x}) = A \vec{x} $$

$$ B = \left\{ \vec{v_1},, \vec{v_2}, \dots, \vec{v_n} \right\} \quad \mid \quad C = \begin{bmatrix} \vec{v_1} & \vec{v_2} & \dots & \vec{v_n} \end{bmatrix} $$

Where $A$ is the transformation matrix, and $C$ is the change of base matrix. 

Then:

$$ T \begin{pmatrix} \begin{bmatrix} \vec{x} \end{bmatrix}_B \end{pmatrix} = D \begin{bmatrix} \vec{x} \end{bmatrix}_B $$

The transformation matrix $D$ with respect to the basis $B$ is:

$$ D = C^{-1} A C $$

You want to change a basis when a transformation matrix is more computational efficient. 

**Linear Algebra is the art of choosing the right basis.**
