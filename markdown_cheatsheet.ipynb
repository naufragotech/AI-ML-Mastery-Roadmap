# Markdown Cheat Sheet

A quick reference for writing Markdown in Jupyter notebook cells.  
Double-click any cell to see the raw source, then **Shift+Enter** to render.

---
## 1. Headings

```
# Heading 1
## Heading 2
### Heading 3
#### Heading 4
##### Heading 5
###### Heading 6
```

# Heading 1
## Heading 2
### Heading 3
#### Heading 4
##### Heading 5
###### Heading 6

---
## 2. Text Formatting

| What | Syntax | Rendered |
|---|---|---|
| Bold | `**bold**` | **bold** |
| Italic | `*italic*` | *italic* |
| Bold + Italic | `***both***` | ***both*** |
| Strikethrough | `~~strikethrough~~` | ~~strikethrough~~ |
| Inline code | `` `code` `` | `code` |
| Highlight | `<mark>highlight</mark>` | <mark>highlight</mark> |

---
## 3. Line Breaks & Paragraphs

**Line break** — end a line with **two spaces** or use `<br>`:

```
First line  
Second line (two trailing spaces above)

First line<br>Second line
```

First line  
Second line

**New paragraph** — leave a blank line between text:

```
Paragraph one.

Paragraph two.
```

Paragraph one.

Paragraph two.

## 7. Code Blocks

Fenced with triple backticks and an optional language:

````
```python
import numpy as np
v = np.array([1, 2, 3])
```
````

```python
import numpy as np
v = np.array([1, 2, 3])
```

Indent with 4 spaces also works:

```
    x = 42
    print(x)
```

    x = 42
    print(x)

## 8. Tables

```
| Left | Center | Right |
|:-----|:------:|------:|
| a    |   b    |     c |
| d    |   e    |     f |
```

| Left | Center | Right |
|:-----|:------:|------:|
| a    |   b    |     c |
| d    |   e    |     f |

**Alignment:**
- `:---` = left (default)
- `:---:` = center
- `---:` = right

## 10. HTML in Markdown

Jupyter Markdown cells support raw HTML:

```html
<details>
<summary>Click to expand</summary>

Hidden content goes here!

</details>
```

<details>
<summary>Click to expand</summary>

Hidden content goes here!

</details>

---

**Colored text:**

```html
<span style="color:red">Red text</span>
<span style="color:#3498db">Custom blue</span>
```

<span style="color:red">Red text</span> · <span style="color:#3498db">Custom blue</span>

---

**Centered text:**

```html
<center>Centered content</center>
```

<center>Centered content</center>

---
## 11. Escaping Special Characters

Use a backslash `\` to display literal characters:

```
\*not italic\*
\# not a heading
\| not a table
```

\*not italic\*  
\# not a heading  
\| not a table
